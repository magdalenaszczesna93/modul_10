Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkami i wynik swojej pracy wyślij Mentorowi jako Jupyter Noteboook zapisany w Github.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
mu, sigma = 3, 1
x = np.random.normal(mu, sigma, 10000)
y = np.random.normal(mu, sigma, 100)
plt.figure(figsize=(12,8))
plt.plot(x, label='odchylenie standardowe=1, średnia= 3')
plt.plot(y, label='odchylenie standardowe=1, średnia= 3')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.hist(x, bins=200) 
plt.hist(y, bins=20)
plt.show()
```

Wykresy x,y oraz histogramy x,y są do siebie bardzo podobne, jednak na wykresie i histogramie Y widać, że ma mniejszą rozpiętość niż X, co sprawdzimy w dalszej części zakresem.

### Zakres

```python
range_x = max(x) - min(x)
print(range_x)
range_y = max(y) - min(y)
print(range_y)
```

Powyższa hipoteza została potwierdzona.

### Średnia

```python
mean_x = np.round(np.mean(x))
mean_y = np.round(np.mean(y))
print(f'mean_a: {mean_x}, mean_b: {mean_y}')
```

```python
plt.hist(x, bins=300)
plt.hist(y, bins=300)
plt.axvline(x = mean_x, color='red', label='Średnia')
plt.axvline(x = mean_y, color='orange', label='Średnia')
plt.legend(loc='upper right')
plt.show()
```

Sprawdziliśmy, czy aby napewno nasze dane z rokładu obie mają średnią 3. Na histogramie zatem nakładają się na siebie.

### Mediana

```python
median_x = np.median(x)
median_y = np.median(y)
print(median_x)
print(median_y)
```

```python
plt.figure(figsize=(12,8))
plt.hist(x, bins=500)
plt.hist(y, bins=500)
plt.axvline(x=mean_x, color='red', label='Średnia x/y')
plt.axvline(x=median_x, color='green', label='Mediana_x')
plt.axvline(x=median_y, color='orange', label='Mediana_y')
plt.legend(loc='upper right')
plt.show()
```

Na histogramie, jak i w obliczeniach można zauważyć, że mediana X jest nieznacznie większa od średniej, natomiast w przypadku mediany Y jest już ona bardziej widocznie mneijsza od średniej. 

### Moda (Dominanta)

```python
mode_x = stats.mode(x)
print(mode_x)
mode_y = stats.mode(y)
print(mode_y)
```

```python
plt.figure(figsize=(12,8))
plt.hist(x, bins=500)
plt.hist(y, bins=500)
plt.axvline(x=mean_x, color='red', label='Średnia x/y')
plt.axvline(x=median_x, color='green', label='Mediana x')
plt.axvline(x=median_y, color='orange', label='Mediana y')
plt.axvline(x=(mode_x[0]), color='grey', label='Moda x')
plt.axvline(x=(mode_y[0]), color='black', label='Moda y')
plt.legend(loc='upper right')
plt.show()
```

Wartośc dominanty X jest poniżej zera, jest mniejsza niż średnia. Obie dominanty są na lewej krawędzi histogramu, dośc oddalone od średniej i median.

### Kwartyle

```python
qx0 = np.quantile(x, 0.0)
qx1 = np.round(np.quantile(x, 0.25))
qx3 = np.round(np.quantile(x, 0.75))
qx4 = np.round(np.quantile(x, 1.0))
print(f'Qx0: {qx0}')
print(f'Qx1: {qx1}')
print(f'Qx2: {median_x}')
print(f'Qx3: {qx3}')
print(f'Qx4: {qx4}'+'\n')

qy0 = np.quantile(y, 0.0)
qy1 = np.round(np.quantile(y, 0.25))
qy3 = np.round(np.quantile(y, 0.75))
qy4 = np.round(np.quantile(y, 1.0))
print(f'Qy0: {qy0}')
print(f'Qy1: {qy1}')
print(f'Qy2: {median_y}')
print(f'Qy3: {qy3}')
print(f'Qy4: {qy4}')
```

```python
fig, axes = plt.subplots(figsize=(15,8), nrows=1, ncols=2)
axes[0].hist(x, bins=500)
axes[0].axvline(x=mean_x, color='red', label='Średnia')
axes[0].axvline(x=median_x, color='green', label='Mediana/Q2')
axes[0].axvline(x=(mode_x[0]), color='orange', label='Moda x')
axes[0].axvline(x=qx0, color='black', label='Min')
axes[0].axvline(x=qx1, color='black', label='Q1')
axes[0].axvline(x=qx3, color='black', label='Q3')
axes[0].axvline(x=qx4, color='black', label='Max')
axes[1].hist(y, bins=50)
axes[1].axvline(x=mean_y, color='red', label='Średnia')
axes[1].axvline(x=median_y, color='green', label='Mediana/Q2')
axes[1].axvline(x=(mode_y[0]), color='orange', label='Moda y')
axes[1].axvline(x=qy0, color='black', label='Min')
axes[1].axvline(x=qy1, color='black', label='Q1')
axes[1].axvline(x=qy3, color='black', label='Q3')
axes[1].axvline(x=qy4, color='black', label='Max')
axes[1].legend(loc='upper right')
plt.show()
```

Oba rozkłady mają pierwszy i trzeci kwartyl w tym samym miejscu, na wartości 2 oraz 4.

### Rozstęp międzykwartylowy - IQR

```python
iqrx = qx3 -qx1
print(iqrx)
iqry = qy3 -qy1
print(iqry)
```

Taki sam rozkład międzykwartylowy wynika z tego, że oba rozkłady mają q1 i q3 w tym samym miejscu.

### Wariancja

```python
variance_x = np.var(x, ddof=1)
print(variance_x)
variance_y = np.var(y, ddof=1)
print(variance_y)
```

### Odchylenie standardowe

```python
print('varx:')
print(variance_x**(1/2))
standard_deviation_x = np.std(x, ddof=1)
print(standard_deviation_x)
print('vary:')
print(variance_y**(1/2))
standard_deviation_y = np.std(y, ddof=1)
print(standard_deviation_y)
```

Zarówno wariancja jak i tymbardziej odchylenie standardowe są w obu rozkładach bardzo bliskie 1, ponieważ tak zostały one ustawione przy ich tworzeniu. Obliczenia te potwierdzają, ze rozkłądy zostały poprawnie utworzone. 

### Asymetria rozkładu

```python
skew_x = stats.skew(x)
print(skew_x)
skew_y = stats.skew(y)
print(skew_y)
```

Po policzeniu współczynnika asymetrii, skośność w obu rozkładach wynosi około zera, zatem mamy potwierdzenie, że są to dane o rozkładzie normalnym.
